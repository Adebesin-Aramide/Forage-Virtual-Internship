```python
user_info1 = []

for i in range(len(reviews1)):
    
    reviews_no = reviews1[i]
    
    info =[a.get_text().strip() for a in reviews_no.find_all(class_ ="review-value")]
    user_info1.append(info)
print(user_info1)
```

```python
keys = ["Seat Type", "Route", "Date Flown", "Recommended"]

my_dict1 = {key: [] for key in keys}

for item in user_info1:
    if len(item) == 6:
        for key, value in zip(keys, item[2:6]):
            my_dict1[key].append(value)
            
    elif len(item) == 5:
        for key, value in zip(keys, item[1:5]):
          my_dict1[key].append(value)
          
    elif len(item) == 4:
        for key, value in zip(keys, item):
          my_dict1[key].append(value)  
print(my_dict1)
```

```python
df1 = pd.DataFrame.from_dict(my_dict1)
print(df1.head())
print(df1.shape)
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from textblob import TextBlob
from wordcloud import WordCloud
import nltk
nltk.download("stopwords")
```

```python
british
```

```python
print(british.iloc[0,0])
british["Sentiments"] = british["Sentiments"].apply(nltk.word_tokenize)
print(british.iloc[0,0])
```

```python
stop_words = set(stopwords.words("english"))
```

```python
british["Sentiments"] = british["Sentiments"].apply(lambda x: " ".join([word for word in x if word not in (stop_words)]))
print(british.head(10))
```

```python
def polarity_calc(text):
    try:
        return TextBlob(text).sentiment.polarity
    except:
        return None
    
    
def tag_calc(num):
    if num<0:
        return "Negative"
    elif num>0:
        return "Positive"
    else:
        return "Neutral"
    
british["Polarity"] = british["Sentiments"].apply(polarity_calc)
british["Tag"] = british["Polarity"].apply(tag_calc)

print(british.head(10))
```

```python
(british.groupby("Tag").size()/british["Tag"].count())*100
```
